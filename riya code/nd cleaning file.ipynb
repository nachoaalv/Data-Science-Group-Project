```python
import pandas as pd
import numpy as np
import openpyxl

# Load the Excel file
file_path = (r"C:\Users\riyas\OneDrive - University of Birmingham\Documents\MSc Data Science\2. group project data science\natural disaster cleaning\natural disasters dataset.xlsx")
df = pd.read_excel(file_path)
```

```python
# Load the main dataset
df = pd.read_excel(file_path, sheet_name="EM-DAT Data")
```

```python
missing_percentage = df.isnull().mean() * 100  # Ensure this is calculated correctly
print(missing_percentage)  # Debugging step to verify output

# Drop columns with more than 80% missing values
columns_to_drop = missing_percentage[missing_percentage > 80].index.tolist()
print("Dropping columns:", columns_to_drop)  # Debugging step
df.drop(columns=columns_to_drop, inplace=True)
```

```python
# Fill missing numerical values with mean or 0
num_cols = ['Magnitude', 'Latitude', 'Longitude', 'Total Deaths', 'No. Injured', 
            'No. Affected', 'No. Homeless', 'Total Damage (\'000 US$)', 
            'Total Damage, Adjusted (\'000 US$)']

for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Fill missing categorical values with "Unknown"
cat_cols = ['Event Name', 'Origin', 'Associated Types', 'Admin Units']
for col in cat_cols:
    df[col].fillna("Unknown", inplace=True)
```

```python
df['Entry Date'] = pd.to_datetime(df['Entry Date'], errors='coerce')
df['Last Update'] = pd.to_datetime(df['Last Update'], errors='coerce')
```

```python
# Cap extreme values at 99th percentile
for col in ['Total Deaths', 'No. Injured', 'No. Affected', 'Total Damage (\'000 US$)']:
    upper_limit = df[col].quantile(0.99)
    df[col] = np.where(df[col] > upper_limit, upper_limit, df[col])
```

```python
df.to_csv("cleaned_natural_disasters_2.csv", index=False)
```
