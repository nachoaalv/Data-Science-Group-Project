```python
import pandas as pd

# Load datasets
df_commoncrane = pd.read_csv(r'C:\Users\riyas\OneDrive - University of Birmingham\Documents\MSc Data Science\2. group project data science\cleaned datasets\cleaned_common_crane.csv') 
df_windturbine = pd.read_csv(r'C:\Users\riyas\OneDrive - University of Birmingham\Documents\MSc Data Science\2. group project data science\cleaned datasets\wind_turbine_data.csv') 
```

```python
df_windturbine
```

```python
# Rename columns
df_windturbine.rename(columns={"Start Year": "Year"}, inplace=True)
                      
# Convert column names to uppercase
df_windturbine.columns = [col.upper() for col in df_windturbine.columns]

# Drop the 'CAPACITY' and 'STATUS' columns if they exist
columns_to_drop = ['CAPACITY (MW)', 'STATUS']
df_windturbine = df_windturbine.drop(columns=[col for col in columns_to_drop if col in df_windturbine.columns])

# Display the modified DataFrame
df_windturbine.head()                   
```

```python
df_commoncrane
```

```python
# Drop the 'GLOBAL UNIQUE IDENTIFIER' column if it exists
if 'GLOBAL UNIQUE IDENTIFIER' in df_commoncrane.columns:
    df_commoncrane = df_commoncrane.drop(columns=['GLOBAL UNIQUE IDENTIFIER'])

# Convert 'OBSERVATION DATE' to year in crane dataset
df_commoncrane["OBSERVATION DATE"] = pd.to_datetime(df_commoncrane["OBSERVATION DATE"], dayfirst=True, errors="coerce")
df_commoncrane["Year"] = df_commoncrane["OBSERVATION DATE"].dt.year

# Convert column names to uppercase
df_commoncrane.columns = [col.upper() for col in df_commoncrane.columns]

# Display the modified DataFrame
df_commoncrane.head()
```

```python
df_windturbine.to_csv("wind_turbine.csv", index=False)
```
